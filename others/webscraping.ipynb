```python
from bs4 import BeautifulSoup
import requests
```

```python
url ="https://www.scrapethissite.com/pages/forms/?per_page=100"
```

```python
page = requests.get(url)
```

```python
print(page)
```

```python
soup =BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
print(soup.prettify())
```

```python
soup.find('p',class_ = 'lead').text.strip()
```

```python
soup.find('div', class_ = 'row').text.strip()
```

```python
table = soup.find_all('table')[0]
table
```

```python
table_headers = table.find_all('th')
table_headers
```

```python
table_headers_clean = [headers.text.strip() for headers in table_headers]
table_headers_clean
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=table_headers_clean)
df
```

```python
column_data = table.find_all('td')
column_data
```

```python
column_data_clean = [data.text.strip() for data in column_data]
#another way
# column_data_clean = []
# for data in column_data:
#     clean_data = data.text.strip()
#     column_data_clean.append(clean_data)
column_data_clean    
```

```python
df = pd.DataFrame(columns = table_headers_clean)# reassign the df to avoid duplicates
for i in range(0,len(column_data),len(table_headers_clean)):
    row_data = [column_data[i+j].text.strip() for j in range(len(table_headers_clean))]# column_data[i+j] would access the
    #data at correct index and with respect to correct column
    df.loc[len(df)] = row_data
    
df    

```
