```python
def sum_product(L):
    """L is a list of numbers
    Returns a tuple where firts element is the sum of all elements of L 
    and second element is the product of all elements of L"""
    sum = 0
    product = 1 
    for num in L:
        sum += num
        product *= num
    return (sum,product)
sum_product([1,4,3,5,6,7,65,4])

    
```

```python
def dot_product(tA, tB):
    """
    tA: a tuple of numbers
    tB: a tuple of numbers of the same length as tA
    Assumes tA and tB are the same length.
    Returns a tuple where the:
    * first element is the length of one of the tuples
    * second element is the sum of the pairwise products of tA and tB
    """
    length_tuple = len(tA)
    product = tuple(tA*tB for tA,tB in zip(tA,tB)) 
    sum_of_product = sum(product)

    return (length_tuple,sum_of_product)
a = (1,3,4,5,6)
b = (2,3,4,5,6)
dot_product(a,b)
```

```python

def max_of_both(n, f1, f2) -> float:
    """ n is an int
        f1 and f2 are functions that take in an int and return a float
    Applies f1 and f2 on all numbers between 0 and n (inclusive). 
    Returns the maximum value of all these results.
    """
    maxval = f1(0)
    for num in range(n+1):
        if f1(num) > maxval:
            maxval = f1(num)
        if f2(num) > maxval:
            maxval = f2(num)
    return maxval
print(max_of_both(2, lambda x:x-1, lambda x:x+1)) # prints 3
print(max_of_both(10, lambda x:x*2, lambda x:x/2))  # prints 20            
        

    
```

```python
def log(x,base,epsilon):
     
    """x is an int or float,base is an int,x > 1,
    epsilon > 0 , power >= 1
     Returns float y such that base**y is within epsilon of x
    """
    lower_bound = 1
    while base**lower_bound < x:
        lower_bound += 1
    low = lower_bound - 1
    high = lower_bound + 1
    g = (low + high)/2
    while abs(base**g - x) >= epsilon:
        if base**g < x:
            low = g
        else:
            high = g
        g = (low + high)/2
    return g

log(54321,5,0.001)        
```

```python
def count_sqrts(nums_list):
    """
    nums_list: a list
    Assumes that nums_list only contains positive numbers and that there are no duplicates.
    Returns how many elements in nums_list are exact squares of elements in the same list, including itself.
    """
    count = 0
    list_squared = [e**2 for e in nums_list]
    for e in list_squared:
        if e in nums_list:
            count += 1
    return count

count_sqrts([1,2,3,4,5,25,16,9]) 
       
    
```

```python
def count_matches(d):
    """d is a dictionary
     Returns how many entries have matching keys and values"""
    d_keys = list(d.keys())
    d_values = list(d.values())
    count = 0
    for i in range(len(d_keys)):
        if d_keys[i] in d_values:
            count += 1
    return count
d = {1:2 , 3:4,5:6} 
d1 = {2:2 ,'a':'a' ,6:1}
count_matches(d1)




```

```python
import copy

# Create a self-referential list
L1 = [2]
L1.append(L1)

# Shallow Copy
shallow_copy = L1[:]
print("Shallow Copy:")
try:
    print(shallow_copy)
except RecursionError as e:
    print("Recursion Error with Shallow Copy")

# Deep Copy
try:
    deep_copy = copy.deepcopy(L1)
    print("\nDeep Copy:")
    print(deep_copy)
except Exception as e:
    print("Error with Deep Copy:", e)
```

```python
def gen_primes_range(n):
    primes = [2] # as 2 is the only even prime number
    
    for i in range(3,n,2):
        is_prime = True
        for x in range(3,int(i**0.5)+1,2):
            if i%x == 0:
                is_prime = False
        if is_prime:
            primes.append(i)
    return primes

gen_primes_range(100)            



```
