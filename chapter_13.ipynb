```python
def get_US_pop(filename):
    file = open(filename,'r')
    dates,pop = [],[]

    for i in file:
        data = ''
        for e in i:
            if e in '0123456789 ':
                data += e
        list = data.split(' ')
        dates.append(int(list[0]))
        pop.append(int(list[1]))

    return dates,pop 
           
```

```python
#claude visualization
import matplotlib.pyplot as plt
import numpy as np

# US population data from 1610 to 2010
years = [1610, 1620, 1630, 1640, 1650, 1660, 1670, 1680, 1690, 1700, 1710, 1720, 1730, 1740, 1750, 
         1760, 1770, 1780, 1790, 1800, 1810, 1820, 1830, 1840, 1850, 1860, 1870, 1880, 1890, 1900, 
         1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]

population = [350, 2307, 4646, 26634, 50368, 75058, 111935, 151507, 210037, 250888, 331711, 466185, 
              629293, 905461, 1170760, 1593625, 2148076, 2780400, 3929214, 5308483, 7239881, 9638453, 
              12866020, 17069453, 23191876, 31443321, 38558371, 50155783, 62979766, 76212168, 92228496, 
              106021537, 123202624, 132164569, 151325798, 179323175, 203302031, 226545805, 248709873, 
              281421906, 308745538]

# Set up the figure with multiple subplots
fig = plt.figure(figsize=(15, 10))

# Linear scale plot
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(years, population, 'b-o', linewidth=2, markersize=4)
ax1.set_title('US Population Growth (1610-2010)', fontsize=14)
ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('Population', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.7)

# Format y-axis with millions
millions_formatter = lambda x, pos: f'{x/1e6:.0f}M' if x >= 1e6 else f'{x/1e3:.0f}K' if x >= 1e3 else f'{x:.0f}'
from matplotlib.ticker import FuncFormatter
ax1.yaxis.set_major_formatter(FuncFormatter(millions_formatter))

# Set x-axis ticks
ax1.set_xticks(np.arange(1600, 2050, 50))
ax1.tick_params(axis='x', rotation=45)

# Logarithmic scale plot
ax2 = fig.add_subplot(2, 2, 2)
ax2.semilogy(years, population, 'r-o', linewidth=2, markersize=4)
ax2.set_title('US Population (Log Scale)', fontsize=14)
ax2.set_xlabel('Year', fontsize=12)
ax2.set_ylabel('Population (log scale)', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.set_xticks(np.arange(1600, 2050, 50))
ax2.tick_params(axis='x', rotation=45)

# Growth rate plot
ax3 = fig.add_subplot(2, 1, 2)
growth_rates = []
growth_years = []

for i in range(1, len(population)):
    rate = ((population[i] - population[i-1]) / population[i-1]) * 100
    growth_rates.append(rate)
    growth_years.append(years[i])

ax3.bar(growth_years, growth_rates, color='green', alpha=0.7)
ax3.set_title('US Population Decade Growth Rate (%)', fontsize=14)
ax3.set_xlabel('Decade', fontsize=12)
ax3.set_ylabel('Growth Rate (%)', fontsize=12)
ax3.grid(True, axis='y', linestyle='--', alpha=0.7)
ax3.set_xticks(np.arange(1600, 2050, 50))
ax3.tick_params(axis='x', rotation=45)

# Highlight interesting periods
ax3.annotate('Colonial\nExpansion', xy=(1700, 40), xytext=(1700, 100),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5))

ax3.annotate('Post-WWII\nBaby Boom', xy=(1950, 14), xytext=(1950, 50),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5))

# Set overall layout and title
plt.tight_layout()
fig.suptitle('US Population Analysis (1610-2010)', fontsize=16, y=0.98)
plt.subplots_adjust(top=0.90)

# Save the figure to a file
plt.savefig('us_population_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
def get_country_pop(filename):
    file = open(filename,'r')
    
    pop = []
    for c in file:
            
        list = c.split('\t')
        l = list[2]
        # print(l) 
        pop_str = ''
        for i in l:
            if i in '0123456789':
                pop_str += i
        
        pop.append(int(pop_str))

    return pop
            
```

```python
def c_to_f(c):
    return (c * 9/5) + 32

def city_temp(city):
    """takes the city name and return list of dates and temperatures of that city"""

    
    file = open(r'lec_22_mat\lec25_temperatures.csv','r')

    dates = []
    temp = []

    for e in file:
        list = e.split(',')
        if list[0] == city:
            temp.append(int(c_to_f(float(list[1]))))
            dates.append(list[2])

    return dates,temp
        

   
```

```python
def get_cities():
    file = open(r'lec_22_mat\lec25_temperatures.csv','r')
    cities = []

    for c in file:
        list = c.split(',')
        city = list[0]
        if city not in cities:
            cities.append(city)
    return cities
            
```

```python
cities = get_cities()[1:]
xpts = range(len(cities))
print(cities)
```

```python
def avg_city_temp():
    cities = get_cities()[1:]
    

    avg_temp = []
    city_label = []

    for c in cities:
        dates,temp = city_temp(c)
        avg_temp.append(sum(temp)/len(temp))
        city_label.append(c[0:2])
        # print(c[0:2],sum(temp)/len(temp))   
    return city_label,avg_temp

            
```

```python
def getYearlyTemp(temp,dates,y):

    yearly_temp = []
    
    for i in range(len(temp)):
        if y == dates[i][:4]:
            yearly_temp.append(temp[i])
    return sum(yearly_temp)/len(yearly_temp) , y
        
```

```python
def getcitytempfortheYear(city):
    average__temp = []
    years = []
    dates,temp = city_temp(city)
    for y in range(1961,2016):

        avg_temp ,year = getYearlyTemp(temp,dates,str(y))
        average__temp.append(avg_temp)
        years.append(year)

    return years,average__temp
    
        
```
